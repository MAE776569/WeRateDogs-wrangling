## Clean

```python
twitter_archive_clean = twitter_archive_df.copy()
image_predictions_clean = image_predictions_df.copy()
```

#### Define

- Convert tweets timestamp into datetime by using pd.to_datetime

#### Code

```python
twitter_archive_clean['timestamp'] = pd.to_datetime(twitter_archive_df['timestamp'])
```

#### Test

```python
print(twitter_archive_clean.timestamp.dtype)
```

#### Define

- Remove replyed tweets

#### Code

```python
twitter_archive_clean = twitter_archive_clean[twitter_archive_clean.in_reply_to_status_id.isnull()]
```

#### Test

```python
twitter_archive_clean.in_reply_to_status_id.notnull().any()
```

#### Define

- Remove retweeted tweets

#### Code

```python
twitter_archive_clean = twitter_archive_clean[twitter_archive_clean.retweeted_status_id.isnull()]
```

#### Test

```python
twitter_archive_clean.retweeted_status_id.notnull().any()
```

#### Define

- Drop in_reply_to_status_id, in_reply_to_user_id, source, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp and rating_denominator

#### Code

```python
twitter_archive_clean.drop(columns=['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'rating_denominator'], inplace=True)
```

#### Test

```python
twitter_archive_clean.columns
```

#### Define

- Remove urls from the end of text

#### Code

```python
twitter_archive_clean['text'] = twitter_archive_clean.text.str.strip().replace('(https?:\/\/.*[\r\n]*)', '')
```

#### Test

```python
twitter_archive_clean.text.str.extract('(https?:\/\/.*[\r\n]*)').count()
```

```python
twitter_archive_clean.info()
```

#### Define

- Fix wrong ratings by using regex

#### Code

```python
twitter_archive_clean['rating_numerator'] = twitter_archive_clean.text.str.findall('(\d+)/\d+').str[-1].astype(int)
```

#### Test

```python
np.all(twitter_archive_clean.text.str.findall('(\d+)/\d+').str[-1] == twitter_archive_clean.rating_numerator)
```

```python
twitter_archive_clean.info()
```

#### Define

- Change multiple expanded url to one url only

#### Code

```python
twitter_archive_clean['expanded_urls'] = twitter_archive_clean.expanded_urls.str.split(',', n=1, expand=True)[0]
```

#### Test

```python
twitter_archive_clean.expanded_urls.head()
```

#### Define

- Extract the source value from the anchor tag

#### Code

```python
twitter_archive_clean['source'] = twitter_archive_clean.source.str.extract('<a.*>(.*)</a>')[0]
```

#### Test

```python
twitter_archive_clean.source.unique()
```

#### Define

- Create dog stage column by combining doggo, floofer, pupper and puppo

#### Code

```python
twitter_archive_clean['doggo'].replace('None', '', inplace=True)
twitter_archive_clean['floofer'].replace('None', '', inplace=True)
twitter_archive_clean['pupper'].replace('None', '', inplace=True)
twitter_archive_clean['puppo'].replace('None', '', inplace=True)

twitter_archive_clean['dog_stage'] = (twitter_archive_clean['doggo'] + 
                                    twitter_archive_clean['floofer'] + 
                                    twitter_archive_clean['pupper'] + 
                                    twitter_archive_clean['puppo'])

twitter_archive_clean['dog_stage'].replace('doggopupper', 'doggo pupper', inplace=True)
twitter_archive_clean['dog_stage'].replace('doggopuppo', 'doggo puppo', inplace=True)
twitter_archive_clean['dog_stage'].replace('doggofloofer', 'doggo floofer', inplace=True)

twitter_archive_clean['dog_stage'].replace('', 'unknown', inplace=True)
twitter_archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

#### Test

```python
twitter_archive_clean.dog_stage.value_counts()
```

#### Define

- Replace - in Breads' names with _ and convert it to lower case

#### Code

```python
image_predictions_clean['p1'] = image_predictions_clean.p1.str.lower().str.replace('-', '_')
image_predictions_clean['p2'] = image_predictions_clean.p2.str.lower().str.replace('-', '_')
image_predictions_clean['p3'] = image_predictions_clean.p3.str.lower().str.replace('-', '_')
```

#### Test

```python
print(image_predictions_clean.p1.str.contains('-').any())
print(image_predictions_clean.p2.str.contains('-').any())
print(image_predictions_clean.p3.str.contains('-').any())
```

#### Define

- Create dogs table

#### Code

```python
dogs_df = twitter_archive_clean[['tweet_id', 'name', 'rating_numerator', 'dog_stage']].merge(image_predictions_clean[['tweet_id', 'p1', 'p1_conf', 'p1_dog']], on='tweet_id')
dogs_df['tweet_id'] = dogs_df.tweet_id.astype(str)
```

#### Test

```python
dogs_df.head()
```

#### Define

- Create tweets table

#### Code

```python
tweets_df = twitter_archive_clean[twitter_archive_clean.columns.difference(['name', 'rating_numerator', 'dog_stage'])].merge(api_df, on='tweet_id')
tweets_df['tweet_id'] = tweets_df.tweet_id.astype(str)
```

#### Test

```python
tweets_df.head()
```

```python
#now merge all needed data into one data frame for analysis purposes
df = twitter_archive_clean.merge(image_predictions_clean[['tweet_id', 'p1', 'p1_conf', 'p1_dog']], on='tweet_id').merge(api_df, on='tweet_id')
df['tweet_id'] = df.tweet_id.astype(str)
df.head()
```

```python
#now save all cleaned data
df.to_csv('twitter_archive_master.csv', index=False)
twitter_archive_clean.to_csv('twitter_archive_clean.csv', index=False)
image_predictions_clean.to_csv('image_predictions_clean.csv', index=False)
dogs_df.to_csv('dogs_data.csv', index=False)
tweets_df.to_csv('tweets_data.csv', index=False)
```
